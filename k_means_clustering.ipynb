**KMeans Clustering** is an unsupervised machine learning algorithm used to group similar data points into k distinct clusters based on their features.

 **How it works (Step-by-step):**
* Choose number of clusters (k) by using elbow method(find optimal number of k)

* Initialize k centroids randomly.

* Assign each point to the nearest centroid (based on Euclidean distance).

* Recalculate centroids as the mean of the points in each cluster.

* Repeat steps 3–4 until centroids stop changing (convergence).

**Why use KMeans?**

To find patterns or groups in unlabeled data — like customer segmentation, image compression, anomaly detection, etc.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("/content/income.csv")
df
```

```python
df.info()
```

```python
plt.scatter(df['Age'],df['Income($)'])
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()
```

```python
km = KMeans(n_clusters = 3)
```

```python
y_pred = km.fit_predict(df[['Age', 'Income($)']])
```

```python
print(y_pred)
print(len(y_pred))
```

```python
df['Cluster'] = y_pred
```

```python
df
```

```python
km.cluster_centers_   # centroid location
```

```python
df1 = df[df['Cluster']==0]
df2 = df[df['Cluster']==1]
df3 = df[df['Cluster']==2]
```

```python
plt.scatter(df1['Age'],df1['Income($)'], color = 'green')
plt.scatter(df2['Age'],df2['Income($)'], color = 'red')
plt.scatter(df3['Age'],df3['Income($)'], color = 'blue')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color = 'black', marker = "*", label ='centroid')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Scaling Income
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])

# Scaling Age
scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
```

```python
# Elbow method

sse = [] # minimum sse value will be tthe best k
k_rng = range(1, 10)
for k in k_rng:
  km = KMeans(n_clusters = k)
  km.fit(df[['Age', 'Income($)']])
  sse.append(km.inertia_)
```

**What is inertia_ in KMeans?**

It is the sum of squared distances between each data point and the centroid of its assigned cluster.

Also called within-cluster sum of squares (WCSS).

```python
plt.xlabel("k (number of clusters)")
plt.ylabel("Sum of Squared Error")
plt.plot(k_rng, sse)
plt.show()
```
