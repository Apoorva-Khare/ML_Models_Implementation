# DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

It is a clustering algorithm that groups data points based on their density in a dataset.

**Key Concepts:**
Density-based: DBSCAN groups data points based on how closely packed they are in the data space.

**Epsilon (ε):** This parameter defines the radius of the neighborhood around a data point.

**MinPts:** This parameter specifies the minimum number of data points required within the ε-radius for a point to be considered a core point.



**1. Core Points:**
A point is a core point if at least MinPts points (including itself) are within its ε-radius. These points form the "dense" center of a cluster.

**2. Border Points:**
A point is a border point if it is not a core point but is within the ε-radius of a core point. These points lie on the edge of a cluster.

**3. Noise Points (or Outliers):**
A point is a noise point if it is neither a core point nor a border point. These points are isolated and do not belong to any cluster

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
x, y = make_moons(n_samples = 300, noise = 0.07, random_state = 0)
```

```python
x
```

```python
plt.scatter(x[:,0], x[:,1])
```

```python
db = DBSCAN(eps = 0.2, min_samples = 5)
```

```python
# y_pred = db.fit_predict(x)
labels = db.fit_predict(x)
```

```python
labels
```

```python
plt.scatter(x[:,0],x[:,1], c = labels, cmap = "plasma")
plt.title("DBSCAN Clustering")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

```python
from sklearn.datasets import make_circles
```

```python
x, _ = make_circles(n_samples = 300, factor = 0.5, noise = 0.05, random_state = 0)
```

```python
db = DBSCAN(eps = 0.2, min_samples = 5)
labels = db.fit_predict(x)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x[:,0],x[:,1], c = labels, cmap = "rainbow")
plt.title("DBSCAN Clustering make circle")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```
